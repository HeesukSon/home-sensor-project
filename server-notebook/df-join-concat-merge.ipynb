# Concatenate DataFrames
Reference webpage (<a href="https://www.datacamp.com/community/tutorials/joining-dataframes-pandas">link</a>)

```python
dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}
dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
```

```python
df1 = pd.DataFrame(dummy_data1)
df2 = pd.DataFrame(dummy_data2)
df3 = pd.DataFrame(dummy_data3)
```

```python
df1, df2, df3
```

```python
df_row = pd.concat([df1, df2])
df_row
```

```python
df_row_reindex = pd.concat([df1, df2], ignore_index=True)
df_row_reindex
```

```python
df_keys = pd.concat([df1, df2], keys = ['x', 'y'])
df_keys
```

```python
df_keys.loc['y']
```

```python
df_col = pd.concat([df1, df2], axis=1)
df_col
```

# Merge DataFrames

```python
df_merge_col = pd.merge(df_row, df3, on='id')
df_row, df3, df_merge_col
```

# Join DataFrames

## Full Outer Join

```python
df1, df2
```

```python
df_outer = pd.merge(df1, df2, on='id', how='outer')
df_outer
```

## Inner Join
The **INNER JOIN** produces only the set of records that match in both DataFrame A and DataFrame B.

```python
df_inner = pd.merge(df1, df2, on='id', how='inner')
df_inner
```

## Right Join
Produces a complete set of records from the right dataframe while the right side will contain null if there is no match.

```python
df_right = pd.merge(df1, df2, on='id', how='right')
df_right
```

## Left Join

```python
df_left = pd.merge(df1, df2, on='id', how='left')
df_left
```

## Joining on index

```python
df_index = pd.merge(df1, df2, right_index=True, left_index=True)
df_index
```

# Time-series friendly merging

For each row in the left DataFrame, you select the last row in the right DataFrame whose _on_ key is less than the left's key. Both DataFrames must be sorted by the key.

```python
trades = pd.DataFrame({
    'time': pd.to_datetime(['20160525 13:30:00.023',
                            '20160525 13:30:00.038',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.048']),
    'ticker': ['MSFT', 'MSFT','GOOG', 'GOOG', 'AAPL'],
    'price': [51.95, 51.95,720.77, 720.92, 98.00],
    'quantity': [75, 155,100, 100, 100]},
    columns=['time', 'ticker', 'price', 'quantity'])

quotes = pd.DataFrame({
    'time': pd.to_datetime(['20160525 13:30:00.023',
                            '20160525 13:30:00.023',
                            '20160525 13:30:00.030',
                            '20160525 13:30:00.041',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.049',
                            '20160525 13:30:00.072',
                            '20160525 13:30:00.075']),
    'ticker': ['GOOG', 'MSFT', 'MSFT','MSFT', 'GOOG', 'AAPL', 'GOOG','MSFT'],
    'bid': [720.50, 51.95, 51.97, 51.99,720.50, 97.99, 720.50, 52.01],
    'ask': [720.93, 51.96, 51.98, 52.00,720.93, 98.01, 720.88, 52.03]},
    columns=['time', 'ticker', 'bid', 'ask'])
```

```python
trades
```

```python
quotes
```

```python
df_merge_asof = pd.merge_asof(trades, quotes, on='time', by='ticker')
df_merge_asof
```

```python
df_merge_asof_tolerance = pd.merge_asof(trades, quotes, on='time', by='ticker', tolerance=pd.Timedelta('0.01ms'))
df_merge_asof_tolerance
```
